```
import numpy as np 
import matplotlib.pyplot as plt
 
```

```
num_steps=5000
h=0.01
t=np.linspace(0,num_steps*h,num_steps+1)
x=np.zeros(num_steps+1)
v=np.zeros(num_steps+1)
a=5
b=0.14
c=7
x[0]=1
v[0]=0

#verlet steps
#code

for n in np.arange(num_steps):
 if n==0:
   x[1] = x[0] + h*v[0] +  (h**2/2) * ((-b*v[0]-c*x[0]))/a

 else :
   x[n+1] = (1+(h*b)/(2*a))**(-1) * (2*x[n] - x[n-1] + (h*b/(2*a))*x[n-1] - ((h**2*c)/(a))*x[n])


plt.plot(t,x) 
```
